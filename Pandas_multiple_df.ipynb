```python
region_url = 'https://andybek.com/pandas-regions'
schools_url = 'https://andybek.com/pandas-ivies'
liberal_schools_url = 'https://andybek.com/pandas-liberal'
income_url = 'https://andybek.com/pandas-mid'
regions = pd.read_csv(region_url)
schools = pd.read_csv(schools_url)
l_schools = pd.read_csv(liberal_schools_url)
mid_career = pd.read_csv(income_url)
```

```python
def salary_rating(x):
  if x < 10000 :
    return 'Low'
  elif x < 100000:
    return 'Medium'
  else:
    return 'High'
```

```python
import re
def salary_rating_string (x):
  salary = float(re.sub(r'\$|,','',x))
  if salary < 10000 :
    return 'Low'
  elif salary < 100000:
    return 'Medium'
  else:
    return 'High'
```

```python
mid_career.loc[:,'Mid-Career 10th Percentile Salary'].fillna('$0').apply(salary_rating_string)
```
